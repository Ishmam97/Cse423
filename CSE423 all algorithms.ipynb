```python
def findzone(x1,y1,x2,y2):
    dx = x2 - x1
    dy = y2 - y1
    if abs(dx) > abs(dy):
        if dx > 0:
            if dy > 0:
                zone = 0
            else:
                zone = 7
        else:
            if dy > 0:
                zone = 3
            else:
                zone = 4
    else:
        if dx > 0:
            if dy > 0:
                zone = 1
            else:
                zone = 6
        else:
            if dy > 0:
                zone = 2
            else:
                zone = 5
    return dx , dy , zone
def tabulatef(X_,Y_,D,X,Y,chosen):
    df = pd.DataFrame(list(zip(X_, Y_ , D , X , Y, chosen)),columns =['X_', 'Y_' ,'d' ,'X','Y','chosen']) 
    return df
def tabulate(X,Y,D,chosen):
    df = pd.DataFrame(list(zip(X, Y , D , chosen)),columns =['X', 'Y' ,'d' ,'chosen']) 
    return df
    
def convertzone(x1 , y1 , x2 , y2 , zone):
    if zone == 0:
        return x1 , y1 , x2, y2
    if zone == 1:
        x1_ = y1
        x2_ = y2
        y1_ = x1
        y2_ = x2
    if zone == 2:
        x1_ = y1
        x2_ = y2
        y1_ = -x1
        y2_ = -x2
    if zone == 3:
        x1_ = -x1
        x2_ = -x2
        y1_ = y1
        y2_ = y2
    if zone == 4:
        x1_ = -x1
        x2_ = -x2
        y1_ = -y1
        y2_ = -y2
    if zone == 5:
        x1_ = -y1
        x2_ = -y2
        y1_ = -x1
        y2_ = -x2
    if zone == 6:
        x1_ = -y1
        x2_ = -y2
        y1_ = x1
        y2_ = x2
    if zone == 7:
        x1_ = x1
        x2_ = x2
        y1_ = -y1
        y2_ = -y2
    return x1_ , y1_ , x2_ , y2_

def originalzone(x , y, zone):
    if zone == 0:
        return x , y
    if zone == 1:
        x_ = y        
        y_ = x        
    if zone == 2:
        x_ = -y        
        y_ = x        
    if zone == 3:
        x_ = -x        
        y_ = y        
    if zone == 4:
        x_ = -x        
        y_ = -y
    if zone == 5:
        x_ = -y
        y_ = -x        
    if zone == 6:
        x_ = y
        y_ = -x       
    if zone == 7:
        x1_ = x1
        y1_ = -y1        
    return x_ , y_

def midpointalgo(x1, y1, x2, y2 , conv = False):
    dx , dy , zone = findzone(x1, y1 , x2, y2);
    if conv == True:    
        print('Zone conversion')
        x , y , x2_ , y2_ = convertzone(x1 , y1 , x2, y2 , zone)
        dx = x2_ - x
        dy = y2_ - y
        
    else:
        print('No zone conversion')
        x = x1
        y = y1
        x2_ = x2
    
    d = ( 2 * dy ) - dx
    ie = 2 * dy
    ine = 2*(dy-dx)
    
    X_ = []
    Y_ = []
    X = []
    Y = []
    D = []
    chosen = []
    
    X.append(x1)
    Y.append(y1)
    D.append(d)
    
    
    X_.append(x)
    Y_.append(y)
    if conv == False:
        while(x<x2_):        
            if d <= 0:
                chosen.append('E')
                d += ie            
            else:
                chosen.append('NE')
                d+= ine            
                y+=1
            x+=1
            X.append(x)
            Y.append(y)
            D.append(d)

        df = tabulate(X,Y,D,chosen)
        print(df.head(15).to_string(index = False))
    else:
        while(x<x2_):        
            if d <= 0:
                chosen.append('E')
                d += ie            
            else:
                chosen.append('NE')
                d+= ine            
                y+=1
            x+=1
            X_.append(x)            
            Y_.append(y)
            D.append(d)
            x_ , y_ = originalzone(x,y,zone)
            X.append(x_)
            Y.append(y_)
        df = tabulatef(X_,Y_,D,X,Y,chosen)
        print(df.head(20).to_string(index = False))
```

```python
def circlepoints(X ,Y):    
    print("zone 0")
    x6 = Y
    y6 = X
    df = pd.DataFrame(list(zip(x6, y6)),columns =['zone0 X', 'zone 0 Y']) 
    print(df.to_string(index = False))
    print("zone 1")
    x6 = X
    y6 = Y
    df = pd.DataFrame(list(zip(x6, y6)),columns =['zone1 X', 'zone 1 Y']) 
    print(df.to_string(index = False))
    print("zone 2")
    x6 = [(e * -1) for e in X]
    y6 = Y
    df = pd.DataFrame(list(zip(x6, y6)),columns =['zone2 X', 'zone 2 Y']) 
    print(df.to_string(index = False))
    print("zone 3")
    x6 = [(e * -1) for e in Y]
    y6 = X
    df = pd.DataFrame(list(zip(x6, y6)),columns =['zone3 X', 'zone3 Y']) 
    print(df.to_string(index = False))
    print("zone 4")
    x6 = [(e * -1) for e in Y]
    y6 = [(e * -1) for e in X]
    df = pd.DataFrame(list(zip(x6, y6)),columns =['zone4 X', 'zone4 Y']) 
    print(df.to_string(index = False))
    print("zone 5")
    x6 = [(e * -1) for e in X]
    y6 = [(e * -1) for e in Y]
    df = pd.DataFrame(list(zip(x6, y6)),columns =['zone5 X', 'zone5 Y']) 
    print(df.to_string(index = False))
    print("zone 6")
    x6 = X
    y6 = [(e * -1) for e in Y]
    df = pd.DataFrame(list(zip(x6, y6)),columns =['zone6 X', 'zone6 Y']) 
    print(df)
    print("zone 7")
    x6 = Y
    y6 = [(e * -1) for e in X]
    df = pd.DataFrame(list(zip(x6, y6)),columns =['zone7 X', 'zone7 Y']) 
    print(df.to_string(index = False))
    
    
def circle(r , cx = 0 , cy=0):
    d = 1 - r
    x = 0
    y = r
    
    chosen = []
    X = []
    Y = []
    D = []
    X.append(x+cx)
    Y.append(y+cy)
    D.append(d)
    if d<0:
        chosen.append("E")
    else:
        chosen.append("SE")
    while x < y:
        if (d < 0):
            cho = "E"
            d += (2*x) + 3
            x += 1
            X.append(x+cx)
            Y.append(y+cy)
            D.append(d)
            chosen.append("E")
        else:
            d += (2 * x) - (2 * y) + 5
            x+=1
            y-=1
            X.append(x+cx)
            Y.append(y+cy)
            D.append(d)
            chosen.append("SE")
    df = tabulate(X,Y,D,chosen)
    print(df.head(10).to_string(index = False))
    circlepoints(X[:-1],Y[:-1])
```

```python
def dda(x1,y1,x2,y2):
    dx = x2-x1;
    dy = y2-y1;
    
    if abs(dx) > abs(dy):
        steps = round(dx)
    else:
        steps = round(dy)
    
    m = dy/dx;    
    Xinc = dx/steps;
    Yinc = dy/steps;
    x = x1;
    y = y1;
    X = []
    Y = []
    Y_round = []
    pixel = []
    X.append(x1)
    Y.append(y1)
    Y_round.append(round(y))
    pixel.append("( {} , {} )".format(x1,y1))
    for i in range(steps):        
        x+=Xinc;
        y+=Yinc;
        X.append(x)
        Y.append(y)
        Y_round.append(round(y))
        pixel.append("( {} , {} )".format(round(x),round(y)))
    df = pd.DataFrame(list(zip(X[:-1], Y[:-1] , Y_round , pixel)),columns =['X', 'Y' ,'Y round' ,'pixel']) 
    return df
```

```python
import numpy as np
def all_zero(a):
    for i , b in enumerate(a):

        if b!=False:
            break
        else:
            if i==3:
                return True
            continue
    return False

def calc_oc(x , y , limits):
    out = [0,0,0,0] #UDRL
    if x < limits[0]:
        out[3] = 1
    if x > limits[1]:
        out[2] = 1
    if y < limits[2]:
        out[1] = 1
    if y > limits[3]:
        out[0] = 1
    
    return out

def find_intersection(x , y , m , limits):
    
    if x < limits[0]:
        xa = limits[0]
        ya = y +( m * (xa-x))
    if x > limits[1]:
        xa = limits[1]
        ya = y +( m * (xa-x))
    if y < limits[2]:
        ya = limits[2]
        xa = x +( (1 / m) * (ya-y))
    if y > limits[3]:
        ya = limits[3]
        xa = x +( (1 / m) * (ya-y))
    
    return xa , ya

def clip(inp , limits):
    inputs = inp
    oc1 = calc_oc(inputs[0],inputs[1],limits)
    oc2 = calc_oc(inputs[2],inputs[3],limits)
    print("initial oc1 = {} , oc2 = {}".format(oc1,oc2))
    m = (inputs[3]-inputs[1])/(inputs[2]-inputs[0])
    while(True):
        if (oc1 == oc2 == [0,0,0,0]):
            print("output : ({},{}) and ({},{})".format(inputs[0],inputs[1],inputs[2],inputs[3]))
            print("oc1 = {} , oc2 = {} so completely inside".format(oc1,oc2))
            break
        elif (not(all_zero(np.logical_and(oc1,oc2)))):
            print("output : ({},{}) and ({},{})".format(inputs[0],inputs[1],inputs[2],inputs[3]))
            print("oc1 = {} , oc2 = {} so completely outside".format(oc1,oc2))
            break
        else:
            if (not(all_zero(oc1))):
                x1,y1 = find_intersection(inputs[0],inputs[1], m,limits)
                
                print("oc1 before intersection is : {}".format(oc1))
                oc1 = calc_oc(x1,y1,limits)
                print("new x1 , y1 for ({},{}) is ({},{})".format(inputs[0],inputs[1] , x1 , y1))
                print("new oc1 is {}".format(oc1))
                inputs[0] = x1
                inputs[1] = y1
            else:
                x2,y2 = find_intersection(inputs[2],inputs[3],m,limits)
                
                print("oc2 before intersection is : {}".format(oc2))
                oc2 = calc_oc(x2,y2,limits)
                
                print("new x2 , y2 for ({},{}) is ({},{})".format(inputs[2],inputs[3] , x2 , y2))
                print("new oc2 is {}".format(oc2))
                inputs[2] = x2
                inputs[3] = y2
            continue
        
        
```
